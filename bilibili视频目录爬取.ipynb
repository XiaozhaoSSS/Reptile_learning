```python
import requests
from bs4 import BeautifulSoup
import re
```

```python
url='https://www.bilibili.com/video/av30708793/'
#headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36'}
headers = {
        'Accept-Encoding': 'gzip, deflate, br',
        'Accept-Language': 'zh-CN,zh;q=0.9,en;q=0.8,ja;q=0.7',
        'Connection': 'keep-alive',
        'Origin': 'https://www.bilibili.com',
        # 'Referer': 'https://www.bilibili.com/video/' + str(i) + '?spm_id_from=333.338.recommend_report.3',
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36'
    }
raw_cookies="rpdid=|(JJmY~l|kkY0J'ull)|)ulYl; buvid3=C8233F30-6267-43DD-971F-CEB543D2B2AD110267infoc; UM_distinctid=16ac5c1e99585a-05d0b8351e8df-3d644509-1fa400-16ac5c1e996acc; _uuid=45C11606-833D-8F23-FC4D-3648C0EC8DE603087infoc; LIVE_BUVID=AUTO1215611110902153; CURRENT_FNVAL=16; stardustvideo=1; _uuid=33C85F4E-0186-B642-D9D3-34740A2BF04107439infoc; fts=1566980644; sid=8c716bpu; DedeUserID=409495891; DedeUserID__ckMd5=ae2fc4fc7c416a89; SESSDATA=a17f7e20%2C1572077751%2Cee5ff791; bili_jct=0c64b504819b8bc99a7e10ce921953cc"
cookies = {}
for line in raw_cookies.split(';'):
    key, value = line.split('=', 1)
    cookies[key] = value

r=requests.get(url,timeout=15,cookies=cookies,headers =headers)
print(r.status_code)
r.raise_for_status
r.encoding=r.apparent_encoding

soup=BeautifulSoup(r.text,'lxml')
print(soup.find('div',id='playerWrap'))
print(soup.find_all('li',class_='bilibili-player-video-btn-menu-list '))
list_box=soup.find('ul',class_='bilibili-player-video-btn-menu')
print(list_box)
mulus=list_box.find_all('li')
for mulu in mulus:
    curr=mulu.text
    with open('目录.txt','a+') as f:
        f.write(curr,'\n')
```

# 扒下来的内容是空的，把soup转成字符搜索

```python
headers = {
        'Accept-Encoding': 'gzip, deflate, br',
        'Accept-Language': 'zh-CN,zh;q=0.9,en;q=0.8,ja;q=0.7',
        'Connection': 'keep-alive',
        # 'Referer': 'https://www.bilibili.com/video/' + str(i) + '?spm_id_from=333.338.recommend_report.3',
        'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/77.0.3865.90 Safari/537.36'
    }
content=requests.get('https://www.bilibili.com/video/av30708793/', headers=headers, timeout=6).text
content_list=content.split('"part"')
mululist=[x for x in content_list if re.match(r'(.*)"duration"(.*),"vid":"","weblink":"","dimension":',x)]
mulus=[]
for mulu in mululist:
    m,i=0,0
    while m<2:
        if mulu[i]=='"':
            m+=1
        i+=1
    mulus.append(mulu[2:i-1])
mulus
```
